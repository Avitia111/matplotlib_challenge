```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

combined_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

combined_df
```

```python
# Checking the number of mice.
len(combined_df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df.loc[combined_df.duplicated(subset = ["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.

# Find the duplicate mouse
duplicate_mouse = combined_df.loc[combined_df["Mouse ID"] == "g989"]


duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_dup_df = combined_df[combined_df["Mouse ID"].isin(duplicate_mice)==False]

cleaned_dup_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_dup_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
tumor_mean = cleaned_dup_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
tumor_median = cleaned_dup_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
tumor_var = cleaned_dup_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
tumor_sd = cleaned_dup_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
tumor_sem = cleaned_dup_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]


# This method is the most straighforward, creating multiple series and putting them all together at the end.

summary_stats = pd.DataFrame({
                                "Tumor Mean": tumor_mean,
                                "Tumor Median": tumor_median,
                                "Tumor Variance": tumor_var,
                                "Tumor Standard Dev": tumor_sd,
                                "Tumor Standard Error": tumor_sem
                            })

summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#   and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function

tumor_vol_stats = cleaned_dup_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})

tumor_vol_stats
```

## Quartiles, Outliers and Boxplots

```python
# Start by getting the last (greatest) timepoint for each mouse
# Using reset_index will return the series as a dataframe
best_treatments = cleaned_dup_df.loc[
                                    (cleaned_dup_df["Drug Regimen"] == "Capomulin") |
                                    (cleaned_dup_df["Drug Regimen"] == "Ramicane")|
                                    (cleaned_dup_df["Drug Regimen"] == "Infubinol")|
                                    (cleaned_dup_df["Drug Regimen"] == "Ceftamin")
                                    ]

mouse_group = best_treatments.groupby("Mouse ID")




mouse_df = pd.DataFrame ({"Timepoint": mouse_group["Timepoint"].max()})

```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tum_vol_last_tp = pd.merge(mouse_df, cleaned_dup_df, on=["Mouse ID","Timepoint"],  how="left")
tum_vol_last_tp = tum_vol_last_tp[["Mouse ID", "Timepoint", "Drug Regimen", "Tumor Volume (mm3)"]]
tum_vol_last_tp
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = ["Capomulin", 
              "Ramicane", 
              "Infubinol", 
              "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)

tum_vol_data = []




# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes

    
for treatment in treatments:
    
    drug_data = tum_vol_last_tp.loc[tum_vol_last_tp["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]  
    tum_vol_data.append(drug_data)
    
    quartiles = drug_data.quantile([.25,.50,.75])

    lower_quart = quartiles[.25]
    upper_quart = quartiles[.75]
    iqr = upper_quart-lower_quart
    
    # Determine outliers using upper and lower bounds
    lower_bound = lower_quart - (1.5 * iqr)
    upper_bound = upper_quart + (1.5 * iqr)
    

drug_data
    
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capomulin_data = cleaned_dup_df.loc[cleaned_dup_df["Drug Regimen"] == 'Capomulin']

mouse_number = capomulin_data["Mouse ID"].sample(n=1)
mouse_number.tolist()
cap_mouse = capomulin_data.loc[capomulin_data["Mouse ID"] == mouse_number.tolist()[0]]
cap_mouse

x_axis = cap_mouse["Timepoint"]
volume = cap_mouse["Tumor Volume (mm3)"]
mouse_id = mouse_number.tolist()[0]
plt.plot(x_axis, volume, marker = '^', color = "green", label = mouse_id )
plt.title(f"Mouse {mouse_id} Treated With Capomulin Over Period of Time")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
x_axis = capomulin_data["Weight (g)"]
tumor_vol = capomulin_data["Tumor Volume (mm3)"]

plt.title("Capomulin Tumor Volume Per Mouse Weight")
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume")

plt.scatter(x_axis,tumor_vol, marker = 'o', color = "purple")
plt.show()
```

## Correlation and Regression

```python
mouse_weight = capomulin_data["Weight (g)"]
tumor_volume = capomulin_data["Tumor Volume (mm3)"]

correlation = st.pearsonr(mouse_weight,tumor_volume)
print(f"The correlation is: {round(correlation[0],2)}")

## add a linear regression equation
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, tumor_volume)

regress_values = mouse_weight * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept, 2))
plt.scatter(mouse_weight, tumor_volume)
plt.plot(mouse_weight, regress_values, "r-")
plt.annotate(line_eq, (20,25), fontsize=15, color ="red")
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Size")
plt.show()
```
